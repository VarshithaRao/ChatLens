```python
# re - regular expression
import re 
import pandas as pd
```

```python
f = open('WhatsApp Chat with MCA - III (Batch 2023 - 25).txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
# creating a new df with columns date and message using regex101

pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}(?:\s?[APMapm]{2})?\s?-\s'
```

```python
messages= re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
# creating df
df = pd.DataFrame({'message_date':dates, 'user_message':messages})
# convert message_date type
df['message_date'] = pd.to_datetime(df['message_date'], format = '%d/%m/%Y, %H:%M - ')

df.rename(columns={'message_date': 'date'}, inplace=True)

df.tail()
```

```python
df.shape
```

```python
# separate users and messages
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\w]+?):\s', message)
    if entry[1:]: # username
        users.append(entry[1])
        messages.append(entry[1])
    else:
        users.append('group_notification')
        messages.append(entry[0])
        
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
# extracting year,month,day

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute

df.head()
```
